```julia
input = split.(readlines("19.input"), "")
```

```julia
y = 1
x = findfirst(e -> e == "|", input[1])
dx = 0
dy = 1
s = 0
while true
    if input[y][x] == " "
        break
    elseif input[y][x] == "+"
        if x < length(input[y]) && dx != -1 && input[y][x + 1] != " "
            dx = 1
            dy = 0
        elseif x > 1 && dx != 1 && input[y][x - 1] != " "
            dx = -1
            dy = 0
        elseif y < length(input) && dy != -1 && input[y + 1][x] != " "
            dx = 0
            dy = 1
        elseif y > 1 && dy != 1 && input[y - 1][x] != " "
            dx = 0
            dy = -1
        else
            print("ERROR")
            break
        end
    elseif input[y][x] != "|" && input[y][x] != "-"
        print(input[y][x])
    end
    x += dx
    y += dy
    s += 1
end
s
```
